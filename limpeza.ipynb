```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
#recebendo o csv
df = pd.read_csv('Pokemon.csv',encoding = 'ISO-8859-1')
```

```python
#formato do DataSet
np.shape(df)
```

```python
#colunas usadas
df.columns
```

```python
#Coloca o atributo TOTAL após os atributos básicos
df = df.loc[:,['number', 'name', 'type1', 'type2', 'hp', 'attack', 'defense',
       'sp_attack', 'sp_defense', 'speed', 'total', 'generation', 'legendary']]

#Renomeia as colunas
colunas = {'number':"Número",
           'name':'Nome',
           'type1':'Tipo 1',
           'type2':'Tipo 2',
           'hp':'Vida',
           'attack':'Ataque',
           'defense':'Defesa',
           'sp_attack':'Atq. especial',
           'sp_defense':'Def. especial',
           'speed':'Velocidade',
           'total':'Total',
           'generation':'Geração',
           'legendary':'Lendário?'}
df = df.rename(columns=colunas)
df.head(10)
```

```python
#Checa se há valores nulos
df.isnull().any()
```

```python
#Checa se há valores NaN
df.isna().any()
```

```python
#Conserta a coluna do tipo 2 em casos do pokémon não haver um segundo tipo 
df.fillna({'Tipo 2':df['Tipo 1']},inplace = True)
df.head(10)
```

```python
#Verifica se ainda há valores NaN após a limpeza
df.isna().any()
```

```python
#Média dos atributos
colunas = ['Vida','Ataque','Defesa','Atq. especial','Def. especial','Velocidade',"Total"]
print('Média dos atributos: ')
round(df[colunas].mean(),2)
```

```python
#Contagem de cada tipo de pokemon
Tipos = ['Normal','Fire','Water','Grass','Flying','Fighting','Poison','Electric','Ground','Rock','Ice','Bug','Psychic','Ghost','Steel','Dragon','Dark','Fairy']
for tipo in Tipos:
       filtragem = (df['Tipo 1'] == tipo) | (df['Tipo 2'] == tipo)
       print(f'Há {filtragem.sum()} pokemons com o tipo {tipo}')
```

```python
df.to_csv('Dataframe.csv', index=False, encoding='utf-8')
```
